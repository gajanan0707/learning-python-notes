### Python documentation tools

```python
# dir function is an easy way to grab a list of all the available attributes inside an object
# for example to check what's available inside os module
import os
dir(os)
```

```python
# names that start with '__' are usually interpreter related so to filter them out
available_attr = [x for x in dir(os) if not x.startswith('__')]
print(available_attr)
```

```python
# to see attributes provided in objects of built-in types, run dir on an existing instance of desired type
dir([])

# the same as writing dir(list)
```

```python
dir([]) == dir(list)
```

```python
dir(tuple) == dir(())
```

### Docstrings: __doc__

Docstrings are special comments that provide a description of a function, method, class, or module. They're enclosed in triple quotes (either single '''docstring''' or double """docstring""") in Python, and are used to explain what the code does, its parameters, return values, and any other relevant information. 

Used for high level documentation for a file.

Comments are retained for inspection in __doc__ attributes after the file/module gets imported.

```python
# to see comments attached to modules
print(os.__doc__)
```

```python
# comments for bultin functions
print(f"The difference between pop and remove is: \n\nPop will: {list.pop.__doc__} \n\nOn the other hand, Remove will: {list.remove.__doc__}")
```

```python
print(list.append.__doc__)
```

### PyDoc: help function

PyDoc tool knows how to extract docstrings and format them nicely. Built-in help funtion is the best PyDoc interface as well as PyDoc GUI and web based HTML report.

```python
# to display methods available to modules
import sys
help(sys)
```

```python
help(sys.getsizeof)
```

```python
# to display methods available to built-in types
help(dict)
```
